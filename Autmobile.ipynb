```python
features=['highway-mpg','city-mpg','engine-size','curb-weight','width','length','wheel-base']
target=auto_data['price']
selected_features=auto_data[features]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(selected_features,target,test_size=0.3)
x_train.shape,y_train.shape
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
y_pred=pd.DataFrame(model.predict(x_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
