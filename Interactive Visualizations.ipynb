```python
import plotly
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as offline
offline.init_notebook_mode(connected=True)# for offline visualizations
```

```python
nifty_data=pd.read_csv('nifty.csv')
```

```python
nifty_data.columns
```

```python
nifty_data= nifty_data.drop(['Adj Close','Volume'],axis=1)
```

```python
nifty_data['date']=pd.to_datetime(nifty_data['date'])
```

```python
nifty_data.isna().sum()
```

```python
trace=go.Scatter(x=nifty_data['date'],\
                y=nifty_data['close'],\
                mode='markers',\
                marker=dict( size=7,\
                            color=nifty_data['close'],\
                            colorscale='Rainbow',\
                            showscale=True,\
                            opacity=0.5),
                text=nifty_data['close']
                )
```

```python
data=[trace]
layout = go.Layout(title='Stocks',\
                    hovermode='closest',\
                    xaxis=dict(title='Date'),\
                    yaxis=dict(title='close')

)
```

```python
fig=go.Figure(data=data,layout=layout)
offline.iplot(fig)
```

```python
trace=go.Box(x=nifty_data['date'].dt.month,y=nifty_data['close'])
data=[trace]
offline.iplot(data)
```
