```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('MPOB_data1.csv')
```

```python
data.describe()
```

```python
fig,ax = plt.subplots(figsize=(10,8))

plt.bar('FishOil',data['FishOil '],width=0.5)
plt.bar('Tallow',data['Tallow '],width=0.5)
plt.bar('Lard',data['Lard '],width=0.5)
plt.bar('CoconutOil',data['CoconutOil '],width=0.5)
plt.bar('SunflowerOil',data['SunflowerOil '],width=0.5)
plt.bar('RapeseedOil',data['RapeseedOil '],width=0.5)
plt.bar('GroundnutOil',data['GroundnutOil '],width=0.5)
plt.xlabel('Types of Vegetable seed')
plt.ylabel('Max prices')
plt.title('Price Comparisons of vegetable seed oil')
plt.show()
```
