```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
Data=pd.read_csv("Dataframe.csv")
```

```python
Data.drop(Data.loc[1,:])
```

```python
y=np.array(Data.price).reshape(-1,1)
```

```python
x=np.array(Data.drop('price',axis=1))
```

```python
x.shape,y.shape
```

```python
kf=KFold(n_splits=3)
```

```python
indices=list(kf.split(x))
```

```python
indices[2]
```

```python
i=1

for train,test in kf.split(x):
    print("\nTrain"+str(i),train)
    print("\nTest",test)
    i+=1
```

```python
len(x),len(train),len(test)
```

```python
x_train1,y_train1=x[indices[0][0]],y[indices[0][0]]
x_train2,y_train2=x[indices[1][0]],y[indices[1][0]]

x_test1,y_test1=x[indices[0][1]],y[indices[0][1]]
x_test2,y_test2=x[indices[1][1]],y[indices[1][1]]
```

```python
x_train1.shape
```

```python
x_test1.shape
```

```python
y_train1.shape
```

```python
regressor=LinearRegression()
regressor.fit(x_train1,y_train1)
regressor.score(x_test1,y_test1)
```

```python
regressor.fit(x_train2,y_train2)
regressor.score(x_test2,y_test2)
```

```python
cross_val_score(regressor,x,y,cv=3)
```

```python
indices
```
